```python
import pandas as pd

#load ODI batting data
df = pd.read_csv('Batting/ODI data.csv')

#Look at first 5 rows
df.head()
```

```python
#How many rows and columns?
print("shape:",df.shape)

#Column names
print("\nColumns:", df.columns.tolist())

#data types
print("\nData types:")
print(df.dtypes)

#check missing values
print("\nMissing values:")
print(df.isnull().sum())
```

```python
# Rename messy columns
df = df.rename(columns={
    'Unnamed: 0': 'index',
    'Unnamed: 13': 'extra',
    'Mat': 'matches',
    'Inns': 'innings',
    'NO': 'not_out',
    'Runs': 'runs',
    'HS': 'high_score',
    'Ave': 'average',
    'BF': 'balls_faced',
    'SR': 'strike_rate',
    '100': 'centuries',
    '50': 'fifties',
    '0': 'ducks'
})

# Drop the useless column
df = df.drop(columns=['extra'])

print("Columns after cleaning:")
print(df.columns.tolist())
```

```python
#Basic statistics of the dataset
print("TOtal players in dataset:", len(df))
print("\nTop 5 run scorers in ODIs:")
print(df[['Player','runs','matches','centuries']].head())

print("\nBasic statistics:")
df[['matches','innings']].describe()
```
