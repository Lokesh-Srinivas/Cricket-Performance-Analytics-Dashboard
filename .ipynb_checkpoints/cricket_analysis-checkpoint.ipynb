```python
import pandas as pd

#load ODI batting data
df = pd.read_csv('Batting/ODI data.csv')

#Look at first 5 rows
df.head()
```

```python
#How many rows and columns?
print("shape:",df.shape)

#Column names
print("\nColumns:", df.columns.tolist())

#data types
print("\nData types:")
print(df.dtypes)

#check missing values
print("\nMissing values:")
print(df.isnull().sum())
```

```python
# Rename messy columns
df = df.rename(columns={
    'Unnamed: 0': 'index',
    'Unnamed: 13': 'extra',
    'Mat': 'matches',
    'Inns': 'innings',
    'NO': 'not_out',
    'Runs': 'runs',
    'HS': 'high_score',
    'Ave': 'average',
    'BF': 'balls_faced',
    'SR': 'strike_rate',
    '100': 'centuries',
    '50': 'fifties',
    '0': 'ducks'
})

# Drop the useless column
df = df.drop(columns=['extra'])

print("Columns after cleaning:")
print(df.columns.tolist())
```

```python
#Basic statistics of the dataset
print("TOtal players in dataset:", len(df))
print("\nTop 5 run scorers in ODIs:")
print(df[['Player','runs','matches','centuries']].head())

print("\nBasic statistics:")
df[['matches','innings']].describe()
```

```python
#Load all three formats of the game
odi = pd.read_csv('Batting/ODI data.csv')
t20 = pd.read_csv('Batting/t20.csv')  
test = pd.read_csv('Batting/test.csv')

print("ODI players:", len(odi))
print("T20 players:", len(t20))
print("Test players:", len(test))
```

```python
# Function to clean any batting dataframe
def clean_batting(df):
    df = df.rename(columns={
        'Unnamed: 0': 'index',
        'Unnamed: 13': 'extra',
        'Mat': 'matches',
        'Inns': 'innings',
        'NO': 'not_out',
        'Runs': 'runs',
        'HS': 'high_score',
        'Ave': 'average',
        'BF': 'balls_faced',
        'SR': 'strike_rate',
        '100': 'centuries',
        '50': 'fifties',
        '0': 'ducks'
    })
    if 'extra' in df.columns:
        df = df.drop(columns=['extra'])
    return df

# Apply cleaning to all three
odi = clean_batting(odi)
t20 = clean_batting(t20)
test = clean_batting(test)

print("All three datasets cleaned!")
print("ODI columns:", odi.columns.tolist())
```

```python
# Convert runs to numeric so we can sort properly
odi['runs'] = pd.to_numeric(odi['runs'], errors='coerce')
t20['runs'] = pd.to_numeric(t20['runs'], errors='coerce')
test['runs'] = pd.to_numeric(test['runs'], errors='coerce')

# Top 10 run scorers in each format
print(" Top 10 ODI Run Scorers:")
print(odi[['Player', 'runs', 'matches', 'centuries']].sort_values('runs', ascending=False).head(10).to_string())

print("\n Top 10 T20 Run Scorers:")
print(t20[['Player', 'runs', 'matches', 'centuries']].sort_values('runs', ascending=False).head(10).to_string())

print("\n Top 10 Test Run Scorers:")
print(test[['Player', 'runs', 'matches', 'centuries']].sort_values('runs', ascending=False).head(10).to_string())
```

```python
# Convert average to numeric
odi['average'] = pd.to_numeric(odi['average'], errors='coerce')
test['average'] = pd.to_numeric(test['average'], errors='coerce')

# Best batting averages in ODIs (minimum 50 matches)
print(" Best Batting Averages in ODIs (min 50 matches):")
print(odi[odi['matches'] >= 50][['Player', 'average', 'matches', 'runs']]
      .sort_values('average', ascending=False).head(10).to_string())

print("\n Best Batting Averages in Tests (min 50 matches):")
print(test[test['matches'] >= 50][['Player', 'average', 'matches', 'runs']]
      .sort_values('average', ascending=False).head(10).to_string())
```

```python
# Columns that should be numeric
numeric_cols = ['matches', 'innings', 'not_out', 'runs', 
                'average', 'balls_faced', 'strike_rate', 
                'centuries', 'fifties', 'ducks']

# Convert for all three datasets
for col in numeric_cols:
    if col in odi.columns:
        odi[col] = pd.to_numeric(odi[col], errors='coerce')
    if col in t20.columns:
        t20[col] = pd.to_numeric(t20[col], errors='coerce')
    if col in test.columns:
        test[col] = pd.to_numeric(test[col], errors='coerce')

print("✅ All numeric columns converted successfully!")
print("\nODI data types:")
print(odi.dtypes)
```

```python
# Drop the index column (not needed)
odi = odi.drop(columns=['index'], errors='ignore')
t20 = t20.drop(columns=['index'], errors='ignore')
test = test.drop(columns=['index'], errors='ignore')

# Remove duplicate players
odi = odi.drop_duplicates(subset=['Player'])
t20 = t20.drop_duplicates(subset=['Player'])
test = test.drop_duplicates(subset=['Player'])

print("✅ Duplicates removed!")
print("ODI players:", len(odi))
print("T20 players:", len(t20))
print("Test players:", len(test))
```

```python
# Add a format column to each dataset
odi['format'] = 'ODI'
t20['format'] = 'T20'
test['format'] = 'Test'

# Save each as a clean CSV file
odi.to_csv('odi_clean.csv', index=False)
t20.to_csv('t20_clean.csv', index=False)
test.to_csv('test_clean.csv', index=False)

# Also save a combined file with all formats together
combined = pd.concat([odi, t20, test], ignore_index=True)
combined.to_csv('cricket_combined.csv', index=False)

print("✅ Clean files saved successfully!")
print("Total combined records:", len(combined))
```

```python
# Load and verify the combined file
df_final = pd.read_csv('cricket_combined.csv')

print("✅ Final dataset verification:")
print("Total records:", len(df_final))
print("Columns:", df_final.columns.tolist())
print("\nRecords per format:")
print(df_final['format'].value_counts())
print("\nSample data:")
df_final.head(3)
```

```python
# Drop unnecessary columns
cols_to_drop = ['4s', '6s', 'Unnamed: 15', 'Unnamed: 11']

odi = odi.drop(columns=cols_to_drop, errors='ignore')
t20 = t20.drop(columns=cols_to_drop, errors='ignore')
test = test.drop(columns=cols_to_drop, errors='ignore')

# Recreate combined and resave
combined = pd.concat([odi, t20, test], ignore_index=True)
combined.to_csv('cricket_combined.csv', index=False)

print("✅ Extra columns dropped!")
print("Final columns:", combined.columns.tolist())
print("Total records:", len(combined))
```

```python
import psycopg2
import pandas as pd
import numpy as np

# Connect to PostgreSQL
conn = psycopg2.connect(
    host="localhost",
    database="cricket_db",
    user="postgres",
    password="Postgre@123"  # your actual password
)

# Load data
df = pd.read_csv('cricket_combined.csv')

# Keep only the columns we need
cols = ['Player', 'Span', 'matches', 'innings', 'not_out', 'runs', 
        'high_score', 'average', 'balls_faced', 'strike_rate', 
        'centuries', 'fifties', 'ducks', 'format']

df = df[cols]

# Replace NaN with None
df = df.where(pd.notnull(df), None)

cur = conn.cursor()

for _, row in df.iterrows():
    cur.execute("""
        INSERT INTO batting_stats 
        (player, span, matches, innings, not_out, runs, high_score, 
         average, balls_faced, strike_rate, centuries, fifties, ducks, format)
        VALUES (%s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s)
    """, tuple(row))

conn.commit()
cur.close()
conn.close()

print("✅ All records loaded into PostgreSQL!")
```
